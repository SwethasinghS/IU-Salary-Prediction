### Decision Tree Model

```python
### max_depth = no of layers in a tree, layer should be upto 5
from sklearn import tree
decisiontree = tree.DecisionTreeRegressor(max_depth = 3)
```

```python
decisiontree.fit(x_train,y_train)
```

### Predict values using trained model

```python
y_train_pred = decisiontree.predict(x_train)
y_test_pred = decisiontree.predict(x_test)
```

```python
y_test_pred
```

```python
y_train_pred
```

### Model Performance of Regression Model

```python
### r2 lies b/w 0 and 1 - 1 is perfect fit & 0 means no fit not find any deviation, 0.4 to 0.8. for good model,
###above 0.8 for excellent model##
from sklearn.metrics import mean_squared_error, r2_score
```

```python
mean_squared_error(y_test,y_test_pred)
```

```python
r2_score(y_train,y_train_pred)
```

```python
r2_score(y_test,y_test_pred)
```
