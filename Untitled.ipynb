```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('C:/Users/Swetha B Singh/Downloads/Salary Prediction/salary_data.csv')
```

```python
dataset.head()
```

```python
features = dataset.iloc[:,1:4].values
```

```python
given_salary = dataset.iloc[:,-1].values
```

```python
index = list(dataset['Sl_No'])
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lreg = LinearRegression()
lreg.fit(features, given_salary)
line_pred_sal = lreg.predict(features)
line_pred_sal
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
polyfeat = PolynomialFeatures()
lreg.fit(polyfeat.fit_transform(features),given_salary)
poly_pred_sal=lreg.predict(polyfeat.fit_transform(features))
poly_pred_sal
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
print('Linear regression mean squared error=', mean_squared_error(given_salary,line_pred_sal))
print('Linear regression accuracy score=',round(r2_score(given_salary,line_pred_sal),2))
print('Polynomial regression mean squared error=',mean_squared_error(given_salary,poly_pred_sal))
print('polynomial regression accuracy score=',round(r2_score(given_salary,poly_pred_sal),2))
```

```python
plt.title('Graph Showing Salary Predicted using Linear Regression')
plt.scatter(index,given_salary,color='grey')
plt.plot(index,line_pred_sal,color='salmon',linewidth=2)
plt.legend(['Linear Regression','Given Salary'],loc='upper left')
plt.xlabel('Index values for each row of feature set')
plt.ylabel('Monthly Salary')
plt.show()
```
